- __Example__: A disease has a survival rate of 30% without treatment. A new treatment is given to 10 patients with the disease.

        - We're looking for $P(x\geq6)$.
            - That's equivalent to $1-P(x\leq5) = 1-.953=.047$
    - b. If we observe 7 patients survive, what can you conclude about the treatment?

        - The probability of 7 people surviving is $0.009$ and we SAW THAT, which is a great thing (it means our treatment works).

## Section 4.5 The Poisson Random Variable
--- 

- Characteristics: 
    1. X = number of times an events occurs per unit of measurement
        - Don't know up front what the $n$ is
        - BUT you do have a time of when the event occurs
    2. The probability an event occurs is the same for all unit
    3. $\#$ of events in one unit independent of $\#$ of events in other units
    4. The expected number of events is $\lambda$ or the MEAN... $E(x) = \mu = \lambda$
    
    
    
- __Example__:
    1. X = number of customers in the drive through line at a particular bank on *Friday morning* (apparently they're busy)
    2. X = number of people that pass through a tollbooth in a day
    3. X = number of traffic accidents per month at University and Alafaya
    
- Probability Distribution:
    - $p(x) = \dfrac {\lambda^x e^{-\lambda}} {x!}$ where $x \in \mathbb{z} \geq 0$
    - Mean and variance: $\mu = \lambda, \sigma^2 = \lambda$
 
- To find Poisson Probabilities:
    1. Point probability formula
    2. Poisson Tables (not in this book - cumulative - on your pink sheets)
    
- Ex. A certain stretch of I-4 has an average of 3 accidents per day and the distribution is poisson. What is the probability of:
    1. Exactly 1 accident today?
        - If you're using your tables... $\lambda = 3, \sigma^2 = 3$
            - Tables read $P(x \leq -)$, so we have to rewrite.
            - $P(x = 1) = P(x\leq1) - P(x\leq0)$
            - $.199 - .05 = .149$
        - If you're using equation...
            - $P(1) = \dfrac {3^1 e^{-3}} {1!} = .14936$

```python
from math import *

def fact(n):
    if n == 0 or n == 1:
        return 1
    f = 1
    for i in range(2, n+1):
        f = f * i
    return f

def poisson(x, l):
    # this calculates P(X <= x)
    return (l ** x) * (e ** (-l)) / (fact(x))

def poissons(xLim, l):
    # this calculates P(X <= xLim)
    sum = 0
    for i in range(0, xLim+1):
        sum += poisson(i, l)
    return sum
```
