## Section 7.5 Determining Sample Size
---

- Allows you to choose a random sample of size $n$ that produces results that are within an acceptable distance from the population value. 

- As long as the sampling is executed properly, excellent estimates result from relatively small samples, no matter how large the population size
    - Researchers have control over how much error there can be.
    - The more precise you want to be, the larger the sample size must be. 
    
- First, you must decide on a degree of precision (specify the sampling error)

- Second, you must set your confidence level (the probability in which the specified precision is achieved). 

### Sample size for estimating means
---

- $n = \dfrac {z^2\sigma^2} {SE^2}$,
    - where $SE =$ sampling error 
    - $\sigma =$ known population standard deviation, or a best guess for it.
    - $z^2 = {z^2}_{\dfrac \alpha 2}$
        - __NOTE__: ALWAYS ROUND UP!!!!
        
        
## Ex. 5
---

-  An estimate is needed of the average age of customers in a department store correct to within 2 years with probability .98.  How many customers should be sampled?  Assume the population standard deviation is 7.5.
    - $n = \dfrac {z^2\sigma^2} {SE^2}$
    - $98\%$ confidence $\rightarrow \alpha = 2\%$
    - We need ${z^2}_{\dfrac \alpha 2} = {z^2}_{\dfrac {0.02} 2} = {z^2}_{0.01} = {2.33}^2$
    - So, $n = \dfrac{ {2.33}^2{7.5}^2 } { {2^2} } = 76.34$
        - $\therefore,\  n = 77$, because we always round up.<br><br>
---
- To decrease the needed sample size $n$, you can: 
    1. Increase the __sampling error__. 
    2. Decrease the __confidence__. 
    3. You have no control over __the standard deviation__.<br><br>
    
### Sample Size for Estimating Proportions
---

- $n = \dfrac {z^2pq} {SE^2}$,
    - where $SE =$ sampling error 
    - $p = $ best guess of p, or p = .5 for largest $n$ necessary
        - __NOTE__: ALWAYS ROUND UP!!!!
        
## Ex. 6
---  

- The Airport Transit Association is considering a poll to determine the percentage of adults who have never flown in an airplane.  How many adults must be surveyed to get a sample percentage with a margin of error of 1.5 percentage points?  Use 95% confidence.
    - $n = \dfrac {{1.96}^2(.5)(.5)} {(0.015)^2}$
    - $\ \ = 4268.4$
    - $\therefore, \ n = 4269$

  
  
## Ex. 7
---
- Suppose they sampled the amount found in # 6 and found that 1280 have never flown.  Find a 95% confidence interval for the percentage of all adults who have never flown.
    - These are proportions<br><br>
    - $\hat p \pm z_{\dfrac \alpha 2} \sqrt{ \dfrac {\hat p \hat q} n}$
    1. $\hat p = \dfrac x n = \dfrac {1280} {4269} = .3$
    2. Check n is large enough
        - $n\hat p \geq 15 \rightarrow 1280 \geq 15$
        - $n\hat q \geq 15 \rightarrow 4265 \geq 15$
        - So, in this case, the central limit theorem applies.
        
    - $\hat p \pm z_{\dfrac \alpha 2} \sqrt{ \dfrac {\hat p \hat q} n} = .3 \pm 1.96\sqrt {\dfrac {.3*.7} {4265} } = .3 \pm .0137$
    - $\therefore,\ $interval is $[.2863, .3137] \rightarrow [28.63\%, 31.37\%]$<br><br>
    
## Considerations is Determining Sample Size
---

1. __Precision__ (the sampling error selected)
2. __Confidence__ (probability selected)
3. __Variability__ in the population (the more variability, the larger sample size needed)
4. __Complexity__ of analysis (the more complex, the larger sample size needed)
5. __Resources__ such as time and money.
