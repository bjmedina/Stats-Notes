## Section 7.2 Continued
---
- __EX__: The NCES surveyed 4400 college graduates abou tthe length of time required to earn their bachelors. THe mean is 5.15 years, std is 1.68 years. __a.__ Construct and interpret a 99% confidence interval for the mean time required by all college graduates. 
    - 4400 is a large enough sample size so that we can use the CLT.
    - Assumptions:
        - None, because of the CLT. 
    - Our interval is $ \bar x \pm Z_{\dfrac \alpha 2} * {\dfrac \sigma {\sqrt{n}}}$<br><br>
    - We have $\bar x$, we have $z_{\dfrac \alpha 2}$, we don't have $\sigma$, but we can use $s$.
    - $ \bar x \pm z_{\dfrac \alpha 2} * {\dfrac \sigma {\sqrt{n}}} = 5.15 \pm 0.065$
    - Interval is $(5.085, 5.215)$
    - "We are $99\%$ confident that the mean time required by all college graduates is between $5.085$ years and $5.215$ years."
        - _ALWAYS_ write this interpretation. <br><br>
        
- __b.__ Now, find 90% confidence interval for the mean time required by all college graduates. Will it be wider or narrower?
    - $ \bar x \pm Z_{\dfrac \alpha 2} * {\dfrac \sigma {\sqrt{n}}} = 5.15 \pm 0.042$
    - Interval is $(5.108, 5.192)$
        - You can get $z_{\dfrac \alpha 2}$ by looking up the associated z value for $\dfrac \alpha 2$
        
### Notes
---
- The higher the confidence, the greater the chance that the CI contains the parameter.
- High confidence coefficients are used in practice so the chance of error is small

### Questions

2. As the confidence coefficient increases, the width __increases__ or __gets wider__. 
3. Is a 95% confidence interval wider or narrower than a 99% confidence interval? __Narrower__


## Section 7.3 Confidence Interval for a Population Mean: Student's t Statistic
---

- Two problems with a small sample:
    1. Shape of the sampling distribution now depends on the population that is sampled. 
        - __Solution__: If the original population is approximately normal, then the sampling distribution is normal. <br><br>
    2. $s$ might not be a good estimator for $\sigma$
        - __Solution__: Use the student t distribution instead of the z distribution (page 817).
        - $t = \dfrac { \bar x - \mu } {\dfrac s {\sqrt n}}$
            - _Looks a LOT like the z-score equation, huh?_
            - _Note_: The t statistic is more variable than the z<br><br>
            
### Characteristics 
---
1. __It's mound shaped.__ 
2. __Symmetric about 0.__ 
3. __More variable than the z.__
4. __There are many t distributions to choose from depending on your degrees of freedom__ (df=n-1).
5. __As $n \rightarrow \infty, t \rightarrow z$__ (that's why the numbers at the bottom of the t table match the critical z values).
6. $t_{\dfrac \alpha 2}$
