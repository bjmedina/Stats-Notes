__To test the efficiency of a procedure designed to increase the daily output of a machine, 5 machines are tested with and without the procedure.__

|  __Results__ | __A__ | __B__  | __C__  | __D__  | __E__  | 
|------|------|------|------|------|------|
| __With__ | $15.6$|   $18.2$ |   $14.3$ |   $14.9$ |  $16.7$| 
| __Without__ | $14.4$ |  $16.7$ |  $13.1$ |   $14.4$ |  $ 14.7$ | 
| __With - Without__ | 1.2 |  1.5 |  1.2 |   0.5 |   2|

<br>

__a. Construct and interpret a 90% Confidence Interval for the difference in daily output with and without the procedure.__ <br><br>

$\bar{x}_D = \dfrac {\sum x} n = \dfrac {6.4} 5 = 1.28$<br><br>
$ {s_D}^2 = \dfrac{ \sum {x^2} - \dfrac{ (\sum x)^2 } n } {n-1} = \dfrac{ 9.38 - \dfrac{ (6.4)^2 } 5 } {5-1} = .297 \rightarrow s_D = .544 $<br><br>
$n_D = 5$<br><br>

__Interval__: $\bar {x}_D \pm t*{\dfrac {s_D} {\sqrt {n_D}}} \rightarrow 1.28 \pm .519 \rightarrow [.761, 1.799]$

> ***NOTE***: We know our work is write because the interval is entirely positive... We did $with\ - \ without$ which gave us positive numbers. So we want positive numbers in our interval.


__Interpretation__: We are $90\%$ the difference in daily output with and without the procedure is between $[.761, 1.799]$

### Paired DIfference Test for $\mu_D$
---

0. __Same as CI__. <br><br>
1. $H_0: \mu_D = D_0.$ $D_0$ __is _usually_ zero.__ <br><br>
2. $H_a: \mu_D \neq D_0$ __or__ $\mu_D > D_0$ __or__ $\mu_D < D_0$<br><br>
3. __Test Statistic__:
> $t = \dfrac { \bar{x}_D - D_0} { S_d \ / \sqrt{n_D}}$
4. __RR__: use t-table based on ($n_d - 1$) degrees of freedom.<br><br>
5. __Conclusion__: Same as always

> $\sqrt { \dfrac{p_1q_1} {n_1} + \dfrac{p_2q_2} {n_2} }$ <br><br>

### Large Sample Confidence Interval for $p_1 - p_2$
---

> $(\hat p_1 - \hat p_2) \pm z * \sqrt { \dfrac{\hat p_1 \hat q_1} {n_1} + \dfrac{\hat p_2 \hat q_2} {n_2} } $<br>

***Assumptions***: 
1. __Independent random samples__
2. __Both samples large (check)__

<br><br>

__Our interval is $[-0.02, 0.22]$, which contains the $0$, which means that the proportions are the same at some point, so you _cant_ conclude that more students oppose a tuition increase that professors__<br><br>
