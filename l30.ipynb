### Large Sample Significance Test for a Mean $\mu$
---

__We use this test in order to determine whether or not we can reject (or not reject) a hypothesis.__<br><br>
__We follow these steps in order.__

0. __Verify the following three__:
   - Requires a random sample of $n \geq 30$, 
   - Variable is quantitative
   - Test is for $\mu$<br><br>
    
1. $H_0:\  \mu = \mu_0$ (or $\mu \leq \mu_0, \ \mu \geq \mu_0$)<br><br>

2. $H_a:\ \mu \neq \mu_0$  ***Two Tailed*** or $H_a:\ \mu > \mu_0$ ***One tailed Upper*** or $H_a:\ \mu < \mu_0$ ***One tailed Lower***
> __NOTE__: You choose this test based on your $H_a$, and you test your $H_0$ against the $Z$ value for $\alpha$ using the test based on $H_a$. See examples below.

3. __Test Statistic__ (summary of sample evidence)
> $z = \dfrac { \bar x - \mu } { \dfrac \sigma {\sqrt n} }$

4. __Rejection Region__ (from $H_a$):
    - __Two Tailed__: $ z < -z_{\dfrac \alpha 2}$ or $ z > z_{\dfrac \alpha 2}$
    - __One tailed__: 
        - _Upper_: $z > z_{\alpha}$
        - _Lower_: $z < -z_{\alpha}$<br><br>
    
5. __Conclusion in terms of the problem__


### Rejection Regions for common values of $\alpha$
---

- For __Lower Tail Test__, put $\alpha$ in lower tail of distribution, fill in area to the left of $-\alpha$ ,and find the $Z$ value.
- For __Upper Tail Test__, put $\alpha$ in upper tail, fill in area to the right of $\alpha$, and find associated $Z$ value.
- For __2 Tailed Test__, put half of $\alpha$ on both sides of distributions, fill in area to the left of $-\alpha$ and to the right of $\alpha$, then find associated $Z$ value.

> __NOTE__: Another name for the $Z$ value is the __critical value__.

|  $\alpha$ | Lower | Upper  |  2 Tailed  | 
|------|------|------|------|
| $0.1$ | $-1.28$  |   $1.28$    | $\pm 1.645$
| $0.05$ | $-1.645$  |  $1.645$ | $\pm 1.96$   |
| $0.01$  | $-2.33$ |  $2.33$ |  $\pm 2.575$


```python
###############################################
# Skip this code to look at the pretty pictures
###############################################
# Skip this code to look at the pretty pictures
###############################################

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

'''
Plotting the Normal Distribution
'''
x = np.arange(-3, 3, 0.01)
alpha = -1.5
sig = 1
mu = 0

def y(x):
    return (1 / (sig*((2*np.pi)**(0.5)))) * np.e ** ( -(x-mu)**(2) / (2 * (sig ** 2)))

# I need to make this look nicer lol
prob = y(x)
plt.title("Lower Tail Test")
section = np.arange(-3, -1.5, 1/20.)
plt.fill_between(section,y(section))
plt.plot(x,prob)
plt.xlabel("X variable")
plt.ylabel("Point Probability")
rect = Rectangle((0, 0), 1, 1, fc="b", alpha=0.7)
plt.legend([rect], ["alpha"])
plt.show()

plt.title("Upper Tail Test")
section = np.arange(1.5, 3, 1/20.)
plt.fill_between(section,y(section))
plt.plot(x,prob)
plt.xlabel("X variable")
plt.ylabel("Point Probability")
rect = Rectangle((0, 0), 1, 1, fc="b", alpha=0.7)
plt.legend([rect], ["alpha"])
plt.show()

plt.title("2 Tailed Test")
section1 = np.arange(-3, -2, 1/20.)
section2 = np.arange(2, 3, 1/20.)
plt.fill_between(section1,y(section1))
plt.fill_between(section2,y(section2))
plt.plot(x,prob)
plt.xlabel("X variable")
plt.ylabel("Point Probability")
rect1 = Rectangle((0, 0), 1, 1, fc="b", alpha=0.7)
rect2 = Rectangle((1, 0), 1, 1, fc="#FF4500", alpha=0.7)
plt.legend([rect1, rect2], ["alpha / 2", "alpha / 2"])
plt.show()
```

__A car manufacturer claims that their new economy model gets an average mpg of at least 58.  A 
random sample of 36 of these cars yields a mean of 56 with a standard deviation of 6.  Is their 
evidence to show the claim is too high?  $\alpha = .01$__

0. - Tests $\mu$ (avg. mpg)
   - Variable is quantitative
   - $n = 36$ so the _Central Limit Theorem applies_<br><br>

1. $H_0: \mu \geq 58$. __Why?__ The problem states that "\[a\] car manufacturer claims that their \[car\]... gets an average mpg of at least 58"; this is the _"claim"_. We believe this claim to be true unless evidence otherwise.<br><br>

2. $H_a: \mu < 58$. __Why?__ The problem asks "\[i\]s their evidence to show the claim is too high". You want to check if the claim is too high, so you choose $\mu < 58$, which corresponds to the __Lower Tail Test__.<br><br>

3. __Test Statistic__: $z = \dfrac { \bar x - \mu } { \dfrac \sigma {\sqrt n} } = \dfrac {56 - 58} {\dfrac 6 {\sqrt{36}}} = -2$<br><br>

4. __Rejection Region__: $H_a <$ lower, $\alpha = .01$
    - Compare your answer to your test statistic. Does it fall in your rejection region?
    - In this case, we get $-2.33$ (find the $Z$ value that corresponds to $\alpha = .01$).
    - Is $-2 < -2.33$? 
        - We check $<$ and not $\neq$ and $>$ because $H_a$ told us to use the __Lower Tail Test__. 
        - Our test _FAILS_ to reject $H_0$ since $-2 < -2.33$ is not true.<br><br>
    
5. __Conclusion__: At $\alpha = 0.01$, there is not sufficient evidence to reject $H_0$. There is no evidence to show the claim is too high. 
> __NOTE__: You _never_ accept a hypothesis; you can only reject or not reject it, but _never_ accept.

__A certain nighttime cold medicine bears a label indication the presence of 600 mg of 
acetaminophen in each fluid ounce of the drug.  The FDA randomly selects 65 1 ounce samples and 
finds the mean content is 595 mg, while the standard deviation is 20 mg.  Is there evidence to 
indicate the 600 mg is incorrect?  Use a 5% significance level.__

0. - Tests $\mu$ 
   - _Central Limit Theorem applies_ ($n = 65$) <br><br>
   
1. $H_0: \mu = 600$. __Why?__ This is because the problem says "...cold medicine bears a label indication the presence of 600 mg..."<br><br>

2. $H_a: \mu \neq 600$. __Why?__ Problem asks "Is there evidence to indicate the 600 mg is incorrect?" Based on this, $\neq$ in the hypothesis corresponds to the __Two Tailed Test__.<br><br>

3. __TS__: $z = \dfrac { \bar x - \mu } { \dfrac \sigma {\sqrt n} } = \dfrac {595 - 600} {\dfrac {20} {\sqrt{65}}} = -2.02$<br><br>

4. __RR__: $\alpha = 0.5$
    - Compare your answer to your test statistic. Does it fall in your rejection region?
    - Both tails of the Uniform Distribution can reject $H_0$. 
    - $-2.02$ lies INSIDE the left tail, so we can reject $H_0$<br><br>

5. __Conclusion__: At $\alpha = 0.05$, there is sufficient evidence to reject $H_0$. THere is evidence the 600 mg is incorrect.
