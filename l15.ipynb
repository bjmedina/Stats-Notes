## The Binomial Random Variable
---
- Experiments where only two possible reponses exist (success/failure)


- __Characteristics__:
    1. $n$ identical trials
    2. Only 2 outcomes - success S and failure F
    3. Probability of S remains constant
        - $P(S) = p$
        - $P(F) = 1 - p = q$
    4. Trials are independent
    5. Binomial random variable x is the number of S's in $n$ trials
    
    
- __Example__: 
    - Draw 10 cards with replacement from a deck of 52. Let $x$ = number of hearts in 10. Is $x$ a random variable?
        1. 10 identical trials
        2. 2 Outcomes - heart, not heart
            - P(Heart) remains constant because of the replacement
        3. $p = \dfrac {13} {52}, q = \dfrac {39} {52}$
        4. Independent
        5. $x$ is binomial
    
    
- __...without replacement?__
    1. Not 10 identical trials.
    2. 2 outcomes -- heart, not heart
    3. $p, q$ do no remain constant
    4. Not independent
    5. $x$ is not binomial
       

- __Example__:
    - Let $x$ = number of children in 3 that get sickle cell anemia. Is $x$ a binomial random variable?
        1. 3 identical trials
        2. Only 2 outcomes - get SCA or don't
        3. $p = 1/4, q = 3/4$
        4. Independent
        5. $x$ is binomial
        
        
### To find Binomial Random Variable Probabilities
---
1. Use methods from Ch 3
2. Use Binomial Point Probability Distribution
    - $p(x) = \dfrac {n!} {(n-x)!x!} p^xq^{n-x} = {n \choose x} p^xq^{n-x} $
    - where $x = 0,1,2,...,n$
        - n = number of trials
        - p = prob. of success
        - q = prob. of failure
3. Binomial tables in back of book (pink sheet)
    - EVERYTHING in table reads $P(x\leq \#)$
    - lowest $n=5$...
    
    
- __Example__: Find the probability that x is, given $n=15, p=.3$:
    1. At most 3 (or $P(x \leq 3)$):
        - This is _exactly_ what our table reads
        - $P(x \leq 3)$ = 0.297
    2. At least 3 (or $P(x \geq 3)$):
        - This is _not_ what our tables
        - We have to rewrite this in terms of our complement
            - $P(x \geq 3) = 1 - P(x \leq 2)$
    3. Exactly 3 (or $P(x=3)$):
        - Not what our table reads
        - $P(x=3) = P(x\leq3) - P(x\leq2)$
    4. Less than 3:
        - $P(x < 3) = P(x \leq 2)$
    5. Exceeds 3:
        - $P(x > 3) = 1 - P(x \leq 3)$


```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    f = 1
    for i in range(2, n+1):
        f = f * i
    return f

def c(n, k):
    return fact(n) / (fact(n-k)*fact(k))

def binomial(n, x, p):
    q = 1-p
    return c(n, x) * (p ** x) * (q ** (n-x))
```

### Mean and Variance of a Binomial Random Variable
---

- __Variance__:
