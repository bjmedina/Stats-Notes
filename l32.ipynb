__A tobacco company claims that its best selling cigarettes contain at most 40 mg of nicotine.  A random sample of 15 of these cigarettes contains 38.5 mg with a standard deviation of 3.7 mg.  Is there evidence to contradict the claim? Use $\alpha = .05$__

0. - Variable is Quantitative
   - Test $\mu$
   - $n=15$
   - __Assumptions:__ underlying population is approx. normal.<br><br>
  
1. $H_0: \mu \leq 40$. Claim is at most 40 mg.<br><br>

2. $H_a: \mu > 40$ Contradict the claim.<br><br>

3. __Test Statistic__: $t = \dfrac{ \bar x - \mu } {\dfrac s {\sqrt n}} = \dfrac{38.5-40} {\dfrac {3.7} {\sqrt{15}}}$<br><br>

4. __Rejection Region__: $t \rightarrow df = n-1 = 15-1 = 14$
    - $H_a:\ > upper$. $\alpha = 0.05$
    - We fail to reject<br><br>
    
5. __Conclusion__: At $\alpha = .05$, there is not sufficient evidence to reject $H_0$. There is no evidence to contradict the claim.

__The mean score on a Peace Corps app. test, based on many tests conducted over a period of time, is 80. Fifty perspective applicants have taken a course designed to improve their scores on the test. The scores of the fifty applicants who completed the course had a mean equal to 86.1 and a std deviation of 8.4__
__Do the data provide sufficient evidence to conclude that students taking the course will have a higher mean score than those who do not? Test at the 95% level of confidence__.

0. - Variable is Quantitative
   - Test $\mu$
   - $n = 50$<br><br>
   
1. $H_0: \mu = 80$ because "The mean score on a Peace Corps app. test, __based on many tests__ conducted over a period of time, is 80." <br><br>

2. $H_a: \mu > 80$, which is the __Upper Tail Test__.<br><br>

3. __Test Statistic__: $z = \dfrac { \bar x  - \mu } { \dfrac \sigma {\sqrt n} } = \dfrac {86.1 - 80} {\dfrac {8.4} {\sqrt{50}}} = 5.1349421015$

> __NOTE__: $\bar x$ is the __sample mean__. $\mu$ is the __population mean__.
<br><br>

4. __Rejection Region__: $\alpha = .05$ Is $ 5.1349421015 < 1.645$? __No__<br><br>

4. __P-value__: Need to know __TS__ and __$H_a$__ is >
    - Graph > 5.13, get the p-value fo that region.
    - Is that p-value smaller than $\alpha$.
<br><br>
5. __Conclusion__: At $\alpha = 0.05$, there is enough evidence to reject $H_0$. There is no evidence to contradict the claim.



### Types of Errors
---

__Type I__: Reject $H_0$ when $H_0$ is true (has probability $\alpha$)<br>

__Type II__: Fail to reject $H_0$ when $H_0$ is false (has probability $\beta$)
> ***NOTE***: $\beta$ is usually `impossible` to compute, so we wont do it. So, we do not say "Accept $H_0$" because computing $\beta$ is difficult and usually unknown.
