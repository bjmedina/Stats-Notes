__Patients who contract a rare disease have only a 2% chance of survival, but a new treatment is 
being tested throughout the country.  A total of 800 patients have been treated and 35 have 
survived.  Is the treatment effective?  Use  $\alpha$ = .005__

0. - __Variable__: Survival $\rightarrow$ qualitative
   - Check $n$ large
       - $np_0 \geq 15$ and $nq_0 \geq 15$
       - $800(.02) \geq 15$ and $800(.98) \geq 15$
       - $16 \geq 15$ and $784 \geq 15$
       - So we can conclude that the CLT applies.<br><br>
       
1. $H_0: p = .02$. __Why?__ "...rare disease have only a 2% chance of survival"<br><br>
2. $H_a: p > .02$. __Why?__ We're not doing $\neq$ because that would imply we test $p > .02$ ***and*** $p < 0.02$, and $p < 0.02$ is __NOT__ effective...<br><br>

3. __Test Statistic__: 
> $z = \dfrac{ \hat p - p_0 } { \sqrt {\dfrac {p_0q_0}{n}}} = \dfrac{ .044 - .02 } { \sqrt {\dfrac {0.2*.98}{800}}} = 4.85$<br><br>
> ***NOTE***: $\hat p$ is the population proportion. In this case, it's $\dfrac {35} {800}$<br><br>
---
4. __Rejection Region__: $H_a >, \alpha = 0.005$
    - We reject $H_0$ if our __TS__ $ > z_{\alpha}$
    - $4.85 > 2.575$
    - We reject $H_0$<br><br>
    
OR __P-value__: $H_a >, TS$... Shade region above TS... Find that probability...If $\alpha$ is greater, reject $H_0$
 

    
5. __Conclusion__: At $\alpha = 0.005$, there is sufficient evidence to reject $H_0$. The treatement is effective.

__A survey was conducted to find the proportion of males who favor government assistance for child 
care.  It was found that among 64 males interviewed, 40 favored assistance.  Is there evidence that 
the proportion of males that favor government assistance for child care is different than .60? Use 
$\alpha$ = .05__

0. - __Variable__: favoring government assistance $\rightarrow$ qualitative.
   - Check $n$ large
       - $np_0 \geq 15$ and $nq_0 \geq 15$
       - $64(.6) \geq 15$ and $64(.4) \geq 15$
       - $38.4 \geq 15$ and $25.6 \geq 15$
       - So we can conclude that the CLT applies.<br><br>
       
1. $H_0: p = .6$. <br><br>

2. $H_a: p \neq .6$ __Why__? "Is there evidence that 
the proportion of males that favor government assistance for child care is different than .60? "<br><br>

3. __Test Statistic__: 
> $z = \dfrac{ \hat p - p_0 } { \sqrt {\dfrac {p_0q_0}{n}}} = \dfrac{ .63 - .6 } { \sqrt {\dfrac {0.6*.4}{64}}} = .49$<br><br>
> ***NOTE***: $\hat p$ is the population proportion. In this case, it's $\dfrac {40} {64}$<br><br>

4. __Rejection Region__: $H_a \neq$, $\alpha = 0.05$<br> $z_{\alpha / 2}$ = 1.96... $.49$ is not $>1.96$ __OR__ $<-1.96$, so we fail to reject.<br><br>

5. __Conclusion__: At $\alpha = 0.05$ there is not sufficient evidence to reject $H_0$. No evidence.

## Section 14.2 Single Population Inferences
---

__Sign Test__ - Nonparametric procedure for testing hypotheses about the central limit tendency of a nonnormal probability distribution.<br>

Nonparametric procedures provide inferences about the median, not the mean.<br>

> ***NOTE***: Why the median? The median is less affected by the skewness of the distribution and the presence of outliers.<br>

### Test for median $\eta$
---

0. - Sample is selected randomly from a __continuous probability distribution.__
> $S_B$ = $\#$ of measurements below $\eta_0$<br>$S_A$ = $\#$ of measurements above $\eta_0$<br>
> ***NOTE***: If you have to eliminate measurements, eliminate the ones that are equal to $\eta_0$ __AND__ reduce n<br>
 
1. $H_0: \eta = \eta_0$<br><br>
2. $H_a: \eta <, >, \neq \eta_0$<br><br>
3. __Test Statistic__: 
    - If $H_a$ contains <, $S = S_B$<br>
    - If $H_a$ contains >, $S = S_A$<br>
    - If $H_a$ contains $\neq$, $S = max(S_A, S_B)$<br>
> ***NOTE***: If the $\eta_0$ is _TRULY_ the median, then $S_A = S_B$ 

4. __P-value__: We are counting successes $\rightarrow$ _binomial_
    - Use $x = $ binomial distribution with parameters $n$ and $p=0.5$
    
> ***NOTE***: Why? By the definition of median, we expect half the observations to fall below the median and half to fall above the median (counting success and failures).

- $P-value = P(x \geq S)$ (double for two tailed test)


5. __Conclusion__: Reject $H_0$ if p-value $< \alpha $
