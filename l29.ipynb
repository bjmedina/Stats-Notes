## Section 8.1 Elements of a Hypothesis Test.
---

### STEPS FOR TESTING
---

0. __Assumptions and information__
    - If you skip this and continue, you'll eventually have to start all over... So don't skip this.<br><br>
    
1. __Null Hypotheses__ ($H_0$): "Status quo"; we believe it to be true unless data provide convincing evidence that it is false (claim)

2. __Alternative Hypothesis__ ($H_a$): That which we believe only if the data provide convincing evidence of its truth (It is usually what the researcher is trying to show). 
    
3. __Test Statistic__: Relative standing of sample data; will be a $z, t$, or $S$
    - $S$ is the median test

4. __Rejection Region__ or __p-value__: Values of the test statistic for which we will reject the null hypothesis

5. __Conclusion__ (in terms of the question):
    1. "There is sufficienct evidence to reject $H_0$..."<br>

    2. "There is not sufficient evidence to reject $H_0$..."
    
> ***NOTE***: We __NEVER__ say that there is sufficient data to accept $H_0$


### Types of Errors
---

__Type I__: Reject $H_0$ when $H_0$ is true (has probability $\alpha$)<br>

__Type II__: Fail to reject $H_0$ when $H_0$ is false (has probability $\beta$)
> ***NOTE***: $\beta$ is usually `impossible` to compute, so we wont do it. So, we do not say "Accept $H_0$" because computing $\beta$ is difficult and usually unknown.

---
- __EX__:  `NHC says Hurricane X will hit FL Keys.` 
    - $H_0$: Hurricane X hits the FL Keys.
    - $H_a$: Hurricane X does not hit FL Keys (MISS)
    - We make a table, where the top is the Actual, and the left side are the predictions
        - __Type I__: NHC predicts miss when in actuality it hits
        - __Type II__: NHC predicts hit when in actuality it misses
    
        
| __Predictions\Actual__   | HIT | MISS  |
|------|------|------|
| __Hit__ | Correct  | __B.__ Incorrect (__Type II__)| 
| __Miss__  | __A.__ Incorrect (__TYPE I__)  |  Correct |

> ***A.is a VERY dangerous scenario...***<br>
> ***B. is a VERY annoying scenario...***<br>
---
    
1. __You control the probability of a Type I error by the choice of $\alpha$__<br>
2. __The more serious the consequences of a Type I error, the smaller $\alpha$ should be (new drug testing, for example)__<br>
3. __We will not cover Section 8.7, which is computing $\beta$__<br>
4. __Type I and Type II errors are inversely related - the smaller the $\alpha$, the larger the $\beta$__
5. __To keep both probabilities at a low level, we must select large sample sizes__
6. __In making a decision, we do not know if we have made a Type I or Type II error.__
