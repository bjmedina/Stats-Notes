## Section 6.3 The Central Limit Theorem
---
### Last section before the exam
- Important because the sample mean is used often
- Properties of the Sampling Distribution of $\bar{x}$
    1. Mean of the sampling distribution equals mean of sample population 
        - $\mu_{\bar{X}} = E(\bar{X}) = \mu$
        - mean of the distribution = expected value of the sample mean = whatever the parameter happens to be<br><br>
        
    2. The standard deviation of the sampling distribution is called the standard error of the mean (sampling error) and is
        - $\sigma_{\bar{X}} = \dfrac \sigma {\sqrt{n}}$
        - Describes spread of sampling distribution
        - As n increases, the standard error decreases ( and the sample mean tends to be closer to the population mean)<br><br>
        
    3. For random sampling of _any population_ with mean $\mu$ and standard deviation $\sigma$, as the sample size $n$ grows, the sampling distribution $\bar{X}$ __approaches a normal distribution__ with mean $\mu_{\bar{X}}$ and standard deviation $\sigma_{\bar{X}}$
        - This works with ANY populaiton... _does NOT have to be normally distributed_
        - We will use $n\geq30$ as our sample size necessary for CLT to apply<br><br>
        
- This is very useful since most variables are very skewed or highly discrete

- Now, to change sample means into standard normal variables, you will use: 
    - Before: $z = \dfrac {x-\mu} \sigma$
    - Sampling: $z = \dfrac {\bar{X} - \mu_{\bar{X}}} {\sigma_{\bar{X}}} = \dfrac {\bar{X} - \mu} {\dfrac \sigma {\sqrt{n}}} $
    
    
- __EX__: Suppose samples are taken from the following and the sample mean is calculated for each sample. Describe the sampling distribution of the sample mean $\bar{X}$ for each case.<br><br>

    - a. A sample of size 12 taken from a _standard normal population_
        - Mean of a standard normal population is 0 $N(0,1)$
            - $\mu_{\bar X} = 0$
        - Standard deviation: .29
            - $\sigma_{\bar X} = \dfrac \sigma {\sqrt{n}}$
        - Is this approximately normal? Why or why not?
            - Yes because original population is normally distributed.<br><br>
            
    - b. A sample of size 100 taken from a population with $\mu = 100$ and $\sigma = 15$
        - $\mu_{\bar X} = 20, \sigma_{\bar X} = 1.5 = \dfrac {15} {\sqrt{100}}$
        - Yes, because of CLT<br><br>
        
    - c. Sample size = 15, $\mu = 230, \sigma = 45$
        - $\mu_{\bar X} = 230, \sigma_{\bar X} = \dfrac {45} {\sqrt{15}}$
        - Not approximately normal because $ n = 15 $<br><br>
        
        
- __EX__: Suppose the times to run a mile for 13 year old boys is normally distributed with a mean of 10 minutes and a standard deviation of 1.2 minutes<br><br>

    - a. Find the probability that a randomly chosen 13 year old boy will taken longer than 12 minutes to run a mile
        - $z = \dfrac {x - \mu} \sigma$
        - $P(x \geq 12) = P(z > \dfrac {12 - 10} {1.2}) = P(z > 1.67)$
            - Table doesn't read $P(z > 1.67)$, but it does read $P(z \leq 1.67)$
        - $P(z > 1.67) = .5 - P(x \leq 1.67) = .0475$<br><br>
        
    - b. If 36 of these boys are selected and the average time of the 36 is calulated, find the probability that the mean is less than 9.5 minutes
        - $P(\bar{X} < 9.5) = P(z < \dfrac {9.5 - 10} {\dfrac {1.2} {\sqrt{36}}})$
        - $P(z < -2.5) = 0.5 - P(z \geq -2.5) = 0.0062$<br><br>
