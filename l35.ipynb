# Chapter 9 Comparison of Two Groups
---

__A fundamental analysis in the sciences__<br>
__Used to compare__: _men VS. women_ / _republicans VS. democrats_. <br>
__Compare means for quantitative variables ($\mu_1 - \mu_2$)__ OR __compare proportions for qualitative variables ($p_1 - p_2$)__.<br>
__Sometimes compare control/experimental groups__.<br>

## Section 9.2 Independent Sampling (Large and Small)
---

- __Independent Random Sample__ - where the choice of subjects for one sample does not depend on which subjects are in the other sample _(section 9.2)_
- __Dependent Samples__ - When a sample has repeated observations taken at different times _(ex. before/after administering cholesterol medication -- section 9.3)_<br>

### Properties of the sampling distribution $\bar{x_1} - \bar{x_2}$
---
1. __Mean is ($\mu_1 - \mu_2$)__
2. __Standard error is__:
> $\sqrt{\dfrac {{\sigma_1}^2} {n} + \dfrac {{\sigma_2}^2} {n} }$
3. __Approximately normal for large samples ($n_1 \geq 30$ and $n_2 \geq 30$)__
> ***NOTE***: If one of the samples is less than 30, then you have to do the small sample procedure, BUT we won't have to do that.

### Large Sample CI for ($\mu_1 - \mu_2$)
---
> $\bar{x_1} - \bar{x_2} \pm z*\sqrt{\dfrac {{s_1}^2} {n} + \dfrac {{s_2}^2} {n} }$<br>
> ***NOTE***: Technically, it should be $\sigma$ instead of $s$, but we'll never have $\sigma$, and we can just use $s$ instead. __Why?__ <br>
> __Assumptions:__ We have two independent random samples. _(CLT guarantees sampling distribution is approx. normal and that $s_1$ and $s_2$ are good estimators of $\sigma_1$ and $\sigma_2$)_


#### Example #1
---
__Researchers in obesity wanted to test the effectiveness of dieting with exercise against dieting without exercise.  
Seventy three patients were randomly divided into 2 groups.  Group 1 was put on a program of dieting with 
exercise.  Group 2 dieted only__

|  Results | Group 1 | Group 2  | 
|------|------|------|
| __Mean Loss__ | 27.1 |   16.8   | 
| __STD__ | 3.5  |  5.2 |
| __Sample Size__ | 37 |  36| 

A. __Construct and interpret a 95% confidence interval for the difference in weight loss between these two 
groups__
- $\bar{x_1} - \bar{x_2} \pm z*\sqrt{\dfrac {{s_1}^2} {n} + \dfrac {{s_2}^2} {n} } = (27.1 - 16.8) \pm 1.96*\sqrt{\dfrac {3.5^2} {36} + \dfrac {5.2^2} {36} } = 10.3 \pm 2.04$<br><br>
- We are $95\%$ confident that the different in weight loss is between $8.26$ lbs and $12.36$ lbs

> __Can you conclude that dieting with exercise results in higher weight loss?__<br>

### Notes
---
1. __If this interval contains only positive numbers,  you can conclude that $\mu_1$ is larger than $\mu_2$__<br><br>
2. __If this interval contains only negative numbers, you can conclude that $\mu_2$ is larger than $\mu_1$__<br><br>
3. __If this interval contains zero, there is insufficient evidence to conclude that either $\mu_1$ or $\mu_2$ is larger.__<br><br>


### Large sample Hypothesis Test about $\mu_1 - \mu_2$
---
$\rightarrow$ tests if there is a difference in the means<br>
0. __Assumptions__:
    - Large
    - Independent
    - Random<br><br>
1. $H_0: \mu_1 - \mu_2 = D_0$. __This is the hypothesized difference.__
2. $H_a: \mu_1 - \mu_2$ can either be $> D_0, < D_0,$ or $\neq D_0$. __If you chose Group one to be the larger one, then you'll never get to do the lower tail test.__
3. __Test Statistic__:
> $z = \dfrac{ (\bar{x_1} - \bar{x_2}) - D_0 } {\sqrt{\dfrac {{s_1}^2} {n} + \dfrac {{s_2}^2} {n} }}$
4. __Rejection Region__ or __P-value__: Sample as 1 sample test.
5. __Conclusion__: in terms of the problem

#### Example #1 (Cont.)
---
__Researchers in obesity wanted to test the effectiveness of dieting with exercise against dieting without exercise.  
Seventy three patients were randomly divided into 2 groups.  Group 1 was put on a program of dieting with 
exercise.  Group 2 dieted only__

|  Results | Group 1 | Group 2  | 
|------|------|------|
| __Mean Loss__ | 27.1 |   16.8   | 
| __STD__ | 3.5  |  5.2 |
| __Sample Size__ | 37 |  36| 

B. __Is there evidence at $\alpha = 0.05$ that diet with exercise is more effective in weight loss?__

0. __Assumptions__:
    - Large
    - Independent
    - Random<br><br>
    
1. $H_0: \mu_1 - \mu_2 = 0$ __Why?__ If nothing assumed about the difference, assume 0.
2. $H_a: \mu_1 - \mu_2 > 0$ __Why?__ Diet is more effective.
<br><br>
3. __Test Statistic__:
> $z = \dfrac{ (\bar{x_1} - \bar{x_2}) - D_0 } {\sqrt{\dfrac {{s_1}^2} {n} + \dfrac {{s_2}^2} {n} }} = \dfrac{ (27.1 - 16.8) - 0 } {\sqrt{\dfrac {3.5^2} {37} + \dfrac {5.2^2} {36} }} = 9.90$
4. __Rejection Rejection__: $1.645$ is __MUCH__ smaller than $9.90$...
5. __Conclusion__: At $\alpha = .05$, there is sufficient evidence to reject $H_0$. Yes, there is evidence that dieting with exercise is more effective in weight loss.
