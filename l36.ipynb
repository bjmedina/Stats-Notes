## Section 9.2 Independent Sampling (Large and Small) Cont.
---

### Small Sample CI for ($\mu_1 - \mu_2$)
---
> $\bar{x_1} - \bar{x_2} \pm t*\sqrt{ {s_p}^2 (\dfrac {{s_1}^2} {n_1} + \dfrac {{s_2}^2} {n_2} )}$<br>where
> ${s_p}^2 = \dfrac { (n_1 - 1){s_1}^2 + (n_2 - 1){s_2}^2 } { n_1 + n_2 - 2 }$<br><br>
> ***NOTE***: $t$ is based on ($n_1 + n_2 - 2$) degrees of freedom (${s_p}^2$ is a pooled estimate)


### Assumptions (small sample):
---
1. __Both__ populations are _approximately normal_
2. Population variances __are equal__
3. __Random__, independent samples.<br><br>

__What if the conditions are not met?__
- __If possible__, try selecting _larger samples_
- If above is not possible, run a _nonparametric test_ (Wilcoxon Rank Sum test)


#### Example #2
---

__A new drug to reduce blood pressure is being tested to determine if it is better than one currently used.  A group 
of 20 patients is randomly divided into 2 groups of 10 each.  The reduction of patients taking the new drug has a 
mean of 10 and a standard deviation of .8.  The mean and standard deviation of patients taking the old drug are 
9.1 and .7.__

a. Construct and interpret a $98\%$ confidence interval for the difference in blood pressure reduction.

|  __Data__ | __New__ | __Old__  | 
|------|------|------|
| $\bar x$ | 10|   9.1 | 
| $s$ | 0.8  |  0.7 |
| $n$ | 10 |  10 | 

- Choose New to be your group 1...
- ${s_p}^2 = \dfrac { (n_1 - 1){s_1}^2 + (n_2 - 1){s_2}^2 } { n_1 + n_2 - 2 } = \dfrac { (10 - 1){.8}^2 + (10 - 1){.7}^2 } { 10 + 10 - 2 } = .565$<br><br>
- $(\bar{x_1} - \bar{x_2}) \pm t*\sqrt{ {s_p}^2 (\dfrac {{s_1}^2} {n_1} + \dfrac {{s_2}^2} {n_2} )} \rightarrow (10 - 9.1) \pm 2.552*\sqrt{ 0.565 (\dfrac {{0.8}^2} {10} + \dfrac {{0.7}^2} {10} )} \rightarrow .9 \pm .86$<br><br>

- Our interval is $[0.4, 1.76]$, which is strictly positive... This tells us that $\mu_1 > \mu_2$.<br><br>

- __Conclusion__: We are $98\%$ that the difference in clood pressure reduction is between $[0.4, 1.76]$.

#### Return to the Blood Pressure Example
---

__Is there evidence at $\alpha = 0.1$ that the new drug is better?__

0. Random, Independent, Population, Both Approx<br><br>
1. $H_0: \mu_1 - \mu_2 = 0$ "Same"<br><br>
2. $H_a: \mu_1 - \mu_2 > 0$ "New drug is better than old drug"<br><br>
3. __TS__:
> $t = \dfrac {(\bar{x_1} - \bar{x_2}) - D_0} { \sqrt{ {s_p}^2 (\dfrac 1 {n_1} + \dfrac 1 {n_2} )}} = \dfrac {(10 - 9.1) - 0} { \sqrt{ .565 (\dfrac 1 {10} + \dfrac 1 {10} )}} = 2.68$<br><br>

4. __Reject Region__: $H_a >, \alpha = .01, df = 10 + 10 - 2 = 18... 2.68 > 2.552$ so we reject $H_0$.<br><br>
5. __Conclusion__: At $\alpha = 0.01$, there is sufficience evidence to reject $H_0$. Yes, the new drug is better.<br><br>


## Section 9.3 Paired Difference Experiments
---

- __An experiment in which observations are paired and the differences are analyzed (we are interested in $\mu_D$)__<br><br>
- __Differencing removes sources of variation that tend to inflate $\sigma^2$__<br><br>
- __This method can be used for 2 readings taken on one subject, or for one reading taken on two matched subjects__<br><br>
- Making comparisons within groups of similar experimental units is called blocking<br><br>
- The paired difference experiment is a simple example of a randomized block experiment.<br><br>


### Paired Difference CI for $\mu_D$
---

> $\bar{x_D} \pm t*\dfrac{s_D} {\sqrt{n_D}}$ Where:
> $\bar{x_D} = $ sample mean difference<br>
> $s_D$ = sample standard deviation of differences<br>
> $n_D$ = number of pairs<br>
> $t$ is based on ($n_D - 1$) degrees of freedom<br>

### Assumptions:
---
1. Distribution of the differences is _approximately normal_
2. Sample differences are _randomly selected_ from the __population__.
