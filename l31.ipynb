## Section 8.3 Observed Significance Levels: p-values
---
- __The p-value is the probability of observing a test statistic that is at least as contradictory to the null hypothesis as the value observed.__

> __NOTE__: This is just an alternative way of deciding whether to reject the null hypothesis. This replaces __Step 4__ in our Test Hypothesis for the mean. It's found by using the test statistic and the type of test (one or two tailed).

- __How do we decide what to do with the p-value?__
    - If p-value $< \alpha$, reject $H_0$
    - If p-value $\geq \alpha$, fail to reject $H_0$
    
> __NOTE__: As a researcher, you're always seeking to _reject_ $H_0$; you want what you've found to be the true thing, and you want evidence that goes against the norm and towards your research. __Small p-values indicate stronger evidence against $H_0$__.

We'll revisit the same examples but changing __step 4__. 

### Ex. 1 From Sheet
---

__A car manufacturer claims that their new economy model gets an average mpg of at least 58.  A 
random sample of 36 of these cars yields a mean of 56 with a standard deviation of 6.  Is their 
evidence to show the claim is too high?  $\alpha = .01$__

0. - Tests $\mu$ (avg. mpg)
   - Variable is quantitative
   - $n = 36$ so the _Central Limit Theorem applies_<br><br>

1. $H_0: \mu \geq 58$. __Why?__ The problem states that "\[a\] car manufacturer claims that their \[car\]... gets an average mpg of at least 58"; this is the _"claim"_. We believe this claim to be true unless evidence otherwise.<br><br>

2. $H_a: \mu < 58$. __Why?__ The problem asks if the "evidence to show the claim is too high". You want to check if the claim is too high, so you choose $\mu < 58$, which corresponds to the __Lower Tail Test__.<br><br>

3. __Test Statistic__: $z = \dfrac { \bar x - \mu } { \dfrac \sigma {\sqrt n} } = \dfrac {56 - 58} {\dfrac 6 {\sqrt{36}}} = -2$
---
 
4. __P-value__: (Look at __Test Statistic__, type of test $\rightarrow$ compare to $\alpha$. *Shade where things are more contradictory*.
    - _How do we decide whats more contradictory?_
        - Look at $H_a$, since it's $<$, we shade to the left of $-2$.
        - $p-value = P(x < -2) = .5 - .4772 = 0.0228$
        - $p-value > \alpha = 0.0228 > 0.1 \rightarrow$ We fail to reject $H_0$
---

5. __Conclusion__: At $\alpha = 0.01$, there is not sufficient evidence to reject $H_0$. There is no evidence to show the claim is too high. 


### Ex. 2 From Sheet
---

__A certain nighttime cold medicine bears a label indication the presence of 600 mg of 
acetaminophen in each fluid ounce of the drug.  The FDA randomly selects 65 1 ounce samples and 
finds the mean content is 595 mg, while the standard deviation is 20 mg.  Is there evidence to 
indicate the 600 mg is incorrect?  Use a 5% significance level.__

0. - Tests $\mu$ 
   - _Central Limit Theorem applies_ ($n = 65$) <br><br>
   
1. $H_0: \mu = 600$. __Why?__ This is because the problem says "...cold medicine bears a label indication the presence of 600 mg..."<br><br>

2. $H_a: \mu \neq 600$. __Why?__ Problem asks "Is there evidence to indicate the 600 mg is incorrect?" Based on this, $\neq$ in the hypothesis corresponds to the __Two Tailed Test__.<br><br>

3. __TS__: $z = \dfrac { \bar x - \mu } { \dfrac \sigma {\sqrt n} } = \dfrac {595 - 600} {\dfrac {20} {\sqrt{65}}} = -2.02$
 
---
4. __P-value__: 
    - *This is a __two tailed test__, so we have consider __both__ tails*.
        - $p-value = P(x < -2.02) + P(x > 2.02) = 2*(0.5 - 0.4783) = 0.0434$
        - $p-value < \alpha = 0.0434 < 0.05 \rightarrow$ We reject $H_0$
---

5. __Conclusion__: At $\alpha = 0.05$, there is sufficient evidence to reject $H_0$. There is evidence the 600 mg is incorrect.
