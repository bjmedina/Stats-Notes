__The manatee population of West Central Florida continues to struggle to survive.  Researchers are 
concerned that fewer baby manatees are surviving.  A study completed in 1985 states that the 
median manatee age was 6 years old in 1985.  Last year, a random sample of the area’s manatees 
revealed that 7 of the 10 manatees sampled were older than 6 years old.  Is there evidence the 
manatee population is aging (i.e., the median age exceeds 6 years old)?  Use $\alpha = 0.05$__

0. - Age is continuous
   - $S_a = 7$
   - $S_b = 3$
   - Assume that there are _no_ manatees that are exactly 6 years old because age is continuous.<br><br>
1. $H_0: \eta = 6$<br>
2. $H_a: \eta > 6$<br><br>
3. __Test Statistic__: $S = S_a = 7$<br><br>
4. __P-value__: $P(x \geq 7)$, $bin(n=10, p = .5)$,<br> $P(x \geq 7) = 1 - P(x\leq6) = 1 - .828 = .172$<br> $p-value = .172 > \alpha = .05$<br><br>
5. __Conclusion__: At $\alpha = .05$, there is not sufficient evidence to reject $H_0$. There is no evidence to suggest that the manatee population is aging.

## Formula Review
---

0. __Info and assumptions__<br>
    - __Mean?__
        - $n \geq 30$?
            - No assumptions
        - $n < 30$?
            - Population is approximately normal
    - __Proportion?__ Check $np_0$ and $nq_0$ both $\geq 15$
        - Both true?
            - CLT applies
        - Only one or none?
            - You're shit out of luck
    - __Median?__
        - No assumptions<br><br>
1. __Null Hypothesis__: $\mu, p,$ or $\eta$<br><br>
2. __Alternative Hypothesis__: $\mu, p,$ or $\eta$. Corresponding tests for  $<, >, \neq$<br><br>
3. __Test Statistic__:
    - __Mean?__
        - Large?
            - $z = \dfrac { \bar x - \mu } {\sigma \ / \sqrt{n}}$
        - Small?
            - $df = n-1$
            - $t = \dfrac { \bar x - \mu } {s \ / \sqrt{n}}$<br>
    - __Proportion?__
        - $z = \dfrac{ \hat p - p_0 } { \sqrt {\dfrac {p_0q_0}{n}}}$<br>
    - __Median?__
        - $S = S_a = S_b$
        - If two tailed test, use $max(S_a, S_b)$<br><br>
4. __Rejection Region or P-value__
    - __Mean?__
        - Large Sample?
            - Z-Table
        - Small Sample?
            - T-Table (no p values)
    - __Proportion?__
        - Z-Table
    - __Median?__
        - Binomial table
        - $P(x \geq S') = 1 - P(x \leq (S'-1))$<br><br>
5. __Conclusion__: At $\alpha = \#$, there is/is not sufficient evidence to reject $H_0$. Answer the question.
       
## Errors
---
__Type I__: You reject $H_0$ when $H_0$ is true... $\alpha$<br>
__Type II__: You fail to reject $H_0$ when $H_0$ is false... $\beta$<br>
